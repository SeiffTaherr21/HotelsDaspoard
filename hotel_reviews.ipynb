### ðŸ“‚ Dataset Columns Description

| Column Name | Description |
|------------|-------------|
| Hotel_Address | The physical address/location of the hotel |
| Additional_Number_of_Scoring | Number of additional review sources used to compute rating |
| Review_Date | The date when the review was written |
| Average_Score | Overall average hotel score based on all reviews |
| Hotel_Name | Name of the hotel |

| Positive_Review | Text of positive feedback by the reviewer |
| Review_Total_Positive_Word_Counts | Number of words in the positive review text |
| Total_Number_of_Reviews_Reviewer_Has_Given | Total number of reviews written by the reviewer on the platform |
| Tags | Additional metadata about the stay (room type, trip type, etc.) |
| days_since_review | Number of days since the review was posted |
| lat | Latitude coordinate of the hotel |
| lng | Longitude coordinate of the hotel |
| Reviewer_Score | Score given by the reviewer |

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent="hotel_locator")

geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def get_coordinates(row):
    if pd.notna(row['lat']) and pd.notna(row['lng']):
        return row['lat'], row['lng']
    
    for col in ['Hotel Address', 'City', 'Country']:
        try:
            location = geocode(row[col])
            if location:
                return location.latitude, location.longitude
        except:
            continue
    return np.nan, np.nan

mask_missing = df['lat'].isna() | df['lng'].isna()
df.loc[mask_missing, ['lat', 'lng']] = df[mask_missing].apply(
    lambda row: pd.Series(get_coordinates(row)), axis=1
)

```

```python
df= pd.read_csv('hotel_reviews.csv')
df
```

```python
df["Hotel_Popularity_Score"] = df["Average Score"] * df["Total Number of Reviews"]
```

```python
df['Sentiment Label'] = df.apply(
    lambda row: 'Positive' if row['Review Total Positive Word Counts'] > row['Review Total Negative Word Counts'] 
                else ('Negative' if row['Review Total Positive Word Counts'] < row['Review Total Negative Word Counts'] 
                      else 'Neutral'),
    axis=1
)
```

```python
def get_stay_length_category(nights):
    if nights < 3:
        return "Short Stay"
    elif nights > 7:
        return "Long Stay"
    else:
        return "Medium Stay"

df["Stay_Length_Category"] = df["Nights Stayed"].apply(get_stay_length_category)
```
